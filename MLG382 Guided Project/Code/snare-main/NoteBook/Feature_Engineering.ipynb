```python
def map_gpa_to_gradeclass(gpa):
    if gpa >= 3.5:
        return 0  # A
    elif gpa >= 3.0:
        return 1  # B
    elif gpa >= 2.5:
        return 2  # C
    elif gpa >= 2.0:
        return 3  # D
    else:
        return 4  # F
```

```python
df_new = df.copy()
df_new = df_new.drop(['StudentID','Age','Gender','Ethnicity','ParentalEducation','GPA'], axis=1)
df_new["GradeClass"] = df["GPA"].apply(map_gpa_to_gradeclass)
df_new.head(40)
```

```python
#df_new.describe([x*0.1 for x in range(10)])

Q1 = df_new.quantile(0.25)
Q3 = df_new.quantile(0.75)
IQR = Q3 - Q1

# Define outliers as values outside 1.5 * IQR
outliers = (df_new < (Q1 - 1.5 * IQR)) | (df_new > (Q3 + 1.5 * IQR))

# Get rows that have outliers
outlier_rows = df_new[outliers.any(axis=1)]

# Removes any all rows that contain an outlier.
df_no_outliers = df_new[~outliers.any(axis=1)]

df_no_outliers
```
